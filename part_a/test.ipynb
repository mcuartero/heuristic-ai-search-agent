```python
from search.program import apply_move, apply_eat, apply_cascade
from search.core import Coord, CellState, PlayerColor, Direction, MoveAction, Coord
from search.utils import render_board
```

```python
# Testing apply_move function

board = {
    Coord(3, 3): CellState(PlayerColor.RED, 2),
    Coord(3, 4): CellState(PlayerColor.BLUE, 3),
}

print("Before:")
print(render_board(board))

moved_board = apply_move(board, Coord(3, 3), Direction.Down)

print("After:")
print(render_board(moved_board))
```

```python
# Testing apply_eat function
board = {
    Coord(3, 3): CellState(PlayerColor.RED, 1),
    Coord(3, 4): CellState(PlayerColor.BLUE, 1),
}
print("Before:")
print(render_board(board))

eaten_board = apply_eat(board, Coord(3, 3), Direction.Down)

print("After:")
print(render_board(eaten_board))
```

```python
# Test the apply_cascade function
board = {
    Coord(3, 6): CellState(PlayerColor.RED, 2),
    Coord(3, 7): CellState(PlayerColor.BLUE, 1),
}

print("Before:")
print(render_board(board))

cascaded_board = apply_cascade(board, Coord(3, 6), Direction.Right)

print("After:")
print(render_board(cascaded_board))
```
